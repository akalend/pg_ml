```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/sample_data/Star3642_balanced.csv')
```

```python
for it in df.columns:
  print(it)
```

```python
df.info()
```

```python
df.head(6)
```

```python
y = df['TargetClass']
X = df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75,
                                                    random_state=42)
```

```python
X_test
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier( cat_features=['SpType'] ) # классификатор
model.fit(X_train, y_train) # обучение классификатора
```

```python
predict = model.predict(X_test)
```

```python
pd.DataFrame(predict)
```

```python
model.save_model('star_classification.cbm')
model.save_model('star_classification.json', format='json')
```
